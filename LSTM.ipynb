```python
dataset= pd.concat([true_dataset, fake_dataset], ignore_index=True)
dataset.head(20)
```

```python
#Regular expression function to extract text after "-"
def extract_text_after(text):
    index = text.find('-')
    if index != -1:
        return text[index + len('-'):]
    else:
        return ""
```

```python
dataset['text'] = dataset['text'].apply(lambda x: extract_text_after(x))
dataset.head()
```

```python
dataset['news']=dataset['title']+dataset['text']
dataset.head()
```

```python
X=dataset['news']
y=dataset['label']
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Data preprocessing
def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    # Removing stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    words = [word for word in words if word.isalnum()]
    words = [word for word in words if word not in stop_words]
    # Stemming (optional)
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]
    return ' '.join(words)
```

```python
X=X.apply(preprocess_text)
X.head()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
vocab_size = len(tokenizer.word_index) + 1
```

```python
# Convert text to sequences
sequences = tokenizer.texts_to_sequences(X)
```

```python
# Padding sequences
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='pre')
```

```python
# Splitting data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(padded_sequences,y, test_size=0.2, random_state=42)
```

```python
# LSTM model
model = Sequential()
model.add(Embedding(vocab_size, 100, input_length=max_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Training the model
model.fit(X_train, y_train, epochs=10, batch_size=64)
```

```python
# Evaluating the model
loss_LSTM, accuracy_LSTM = model.evaluate(X_test, y_test)
print("Accuracy:", accuracy_LSTM)
```
